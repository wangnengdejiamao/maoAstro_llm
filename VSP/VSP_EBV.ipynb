```python
import numpy as np
from astropy.coordinates import SkyCoord
from astropy import units as u
from dustmaps.bayestar import BayestarQuery
```

```python
def init_bayestar2019(max_samples=5):
    bayestar = BayestarQuery(max_samples=max_samples, version='bayestar2019')
    return bayestar

def cal_green19(ra,dec,dist,bayestar,mode='median'): 
    # 输入：赤经,赤纬,距离(默认单位deg，deg，pc, 输入纯数字不加单位，可以是包含多个值的向量), 已初始化的Bayestar map, mode可选‘best','median'
    # 功能：返回green19 3D Reddening
    coords = SkyCoord(ra*u.deg, dec*u.deg, distance=dist*u.pc, frame='icrs')
    reddening = bayestar(coords, mode=mode)
    return reddening

def cal_green19_ebv3d(ra,dec,dist,bayestar,mode='median'): 
    # 输入：赤经,赤纬,距离(默认单位deg，deg，pc, 输入纯数字不加单位，可以是包含多个值的向量), 已初始化的Bayestar map, mode可选‘best','median'
    # 功能：返回green19 ebv3d (归一化到SFD98), 输出模式可选中值和最佳
    coords = SkyCoord(ra*u.deg, dec*u.deg, distance=dist*u.pc, frame='icrs')
    reddening = bayestar(coords, mode=mode)
    ebv3d = 0.981*reddening*(3.518-2.617)
    return ebv3d

def cal_green19_ebv3d_pct(ra,dec,dist,bayestar): 
    # 输入：赤经,赤纬,距离(默认单位deg，deg，pc, 输入纯数字不加单位，可以是包含多个值的向量), 已初始化的Bayestar map, mode可选‘best','median'
    # 功能：返回green19 ebv3d (归一化到SFD98), 每一组坐标输出ebv3d的三个分位数
    coords = SkyCoord(ra*u.deg, dec*u.deg, distance=dist*u.pc, frame='icrs')
    reddening = bayestar(coords, mode='percentile', pct=[16., 50., 84.])
    ebv3d = 0.981*reddening*(3.518-2.617)
    return ebv3d
```

```python
print('Initializing bayestar2019...')
bayestar = init_bayestar2019(max_samples=5)
```
