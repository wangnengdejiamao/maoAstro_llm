```python
from VSP_Auxiliary import *
from astropy.table import Table
import lightkurve as lk
import matplotlib.pyplot as plt
import time
# %matplotlib agg
```

```python
def get_lc(coord, lc_path='./', pic_path='./',prefix=''): 
    # 输入坐标对象，lc保存路径，lc图片保存路径，保存文件前缀
    # 函数利用Lightkurve下载Kepler/TESS数据到指定文件
    # 函数返回所有光变曲线数据

    # print('>>> Getting Kepler/TESS data....') 
    # 从lightkurve搜索光变曲线
    try:
        search_result = lk.search_lightcurve(f"{coord.ra.value} {coord.dec.value}", radius=2) # arcsec
    except:
        print('Faild downloading lc from lightcurve! Retrying...')
        time.sleep(2)
        try:
            search_result = lk.search_lightcurve(f"{coord.ra.value} {coord.dec.value}", radius=2) # arcsec
        except:
            time.sleep(5)
            try:
                search_result = lk.search_lightcurve(f"{coord.ra.value} {coord.dec.value}", radius=2) # arcsec
            except:
                search_result = []
    tessn = []
    k2 = 'No'
    kic = 'No'
    for n in range(len(search_result)):
        if 'K2 Campaign' in search_result[n].mission[0]:
            k2='YES'
        if "Kepler Quarter" in search_result[n].mission[0]:
            kic='YES'
        if "TESS" in search_result[n].mission[0]:
            tessn.append(search_result[n].mission[0])
    tess = len(list(set(tessn)))

    print('querying from lightkurve:')
    print('*if Kepler?',kic)
    print('*if K2?    ',k2)
    print('*TESS:     ',tess)
    print('----------------The following data is obtained:----------------')

    if len(search_result)>0:
        lcs = []
        reports = Table(names=['No.','Mission','Year','Author','Cadence'],dtype=['i4','S22','i4','S20','i4'])
        lc_name = prefix
        for n in range(len(search_result)):
            if search_result[n].author[0] == 'EVEREST':
                continue
            if search_result[n].author[0] == 'K2VARCAT':
                continue
            lc_file = lc_path+lc_name+"_%02d_"%(n+1)+search_result[n].mission[0]+'_'+\
                           search_result[n].author[0]+'_'+str(int(search_result[n].exptime[0].value))+'.csv'
            plot_file = pic_path+lc_name+"_%02d_"%(n+1)+search_result[n].mission[0]+'_'+\
                           search_result[n].author[0]+'_'+str(int(search_result[n].exptime[0].value))+'.png'
            reportn = "%2d  "%(n+1)+'%-16s'%search_result[n].mission[0]+' '+\
                  '%-5s'%search_result[n].year[0]+' '+\
                  ' %-10s'%search_result[n].author[0]+' '+\
                  '%6d s'%int(search_result[n].exptime[0].value)
            reportn1 = "No.%02d  "%(n+1)+'%15s'%search_result[n].mission[0]+'  '+\
                  '%9s'%search_result[n].author[0]+'    '+\
                  'year:%5s'%search_result[n].year[0]+'  '+\
                  'Exptime:%5d s'%int(search_result[n].exptime[0].value)
            row = [n+1,search_result[n].mission[0],search_result[n].year[0],search_result[n].author[0],search_result[n].exptime[0].value]
            if 'K2 Campaign' in search_result[n].mission[0]:
                try:
                    lc = search_result[n].download()
                    print(reportn)
                    reports.add_row(row)
                except:
                    try:
                        lc = search_result[n].download()
                        print(reportn)
                        reports.add_row(row)
                    except:
                        try:
                            lc = search_result[n].download()
                            print(reportn)
                            reports.add_row(row)
                        except:
                            print('Error! Could not download:',reportn)
                            continue
                lc.to_csv(lc_file, overwrite=True)
                lcs.append(lc)
                fig, ax = plt.subplots(figsize=(9,3),dpi=300)
                lcplot = lc.plot(ax=ax,c='k',linewidth=.4,alpha=0.9)
                ax.set_title(reportn1, fontsize=8.5)
                ax.legend(loc='upper left', fontsize=8.5)
                fig.savefig(plot_file, bbox_inches = 'tight', dpi=300)
                plt.close(fig)
                plt.close()
            if "Kepler Quarter" in search_result[n].mission[0]:
                try:
                    lc = search_result[n].download()
                    print(reportn)
                    reports.add_row(row)
                except:
                    try:
                        lc = search_result[n].download()
                        print(reportn)
                        reports.add_row(row)
                    except:
                        try:
                            lc = search_result[n].download()
                            print(reportn)
                            reports.add_row(row)
                        except:
                            print('Error! Could not download:',reportn)
                            continue
                lc.to_csv(lc_file, overwrite=True)
                lcs.append(lc)
                fig, ax = plt.subplots(figsize=(9,3),dpi=300)
                lcplot = lc.plot(ax=ax,c='k',linewidth=.4,alpha=0.9)
                ax.set_title(reportn1, fontsize=8.5)
                ax.legend(loc='upper left', fontsize=8.5)
                fig.savefig(plot_file, bbox_inches = 'tight', dpi=300)
                plt.close()
            if "TESS" in search_result[n].mission[0]:
                try:
                    lc = search_result[n].download()
                    print(reportn)
                    reports.add_row(row)
                except:
                    try:
                        lc = search_result[n].download()
                        print(reportn)
                        reports.add_row(row)
                    except:
                        try:
                            lc = search_result[n].download()
                            print(reportn)
                            reports.add_row(row)
                        except:
                            print('Error! Could not download:',reportn)
                            continue
                lc.to_csv(lc_file, overwrite=True)
                lcs.append(lc)
                fig, ax = plt.subplots(figsize=(9,3),dpi=300)
                lcplot = lc.plot(ax=ax,c='k',linewidth=.4,alpha=0.9)
                ax.set_title(reportn1, fontsize=8.5)
                ax.legend(loc='upper left', fontsize=8.5)
                fig.savefig(plot_file, bbox_inches = 'tight', dpi=300)
                plt.close()

                reports.write(lc_path+'lightkurve_query_result.csv',format='ascii.csv',overwrite=True)
    else:
        lcs = None
        reports = None
    print('---------------------------------------------------------------')
    return lcs, reports

# 测试
#c = SkyCoord('06 52 37.19 +24 36 22.1', unit=(u.hourangle, u.deg), frame='fk5')
#create_dir('./Temp/TESS_test/')
#get_lc(c, lc_path='./Temp/TESS_test/', pic_path='./Temp/TESS_test/', prefix='My2')
```
