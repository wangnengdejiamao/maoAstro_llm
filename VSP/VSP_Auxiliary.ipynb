```python
# 导入需要的模块
import wget
import csv
from astropy.io import fits
import os
import numpy as np
from astropy.coordinates import SkyCoord
from astropy import units as u
```

```python
def download_url(url, save_name, overwrite=False, ifprint=False, ifbar=False): 
    # 输入：url下载链接，save_name保存到的文件名称，overwrite为是则删除原有文件后下载(此项打开时，需要谨慎验证保存文件的名称)，
    #      ifprint为是则打印下载状态，ifbar为是则显示下载进度
    # 功能：下载指定链接到对应文件，如果ifprint为True，则打印下载是否完成
    # 若下载成功，返回文件名称；若下载失败，返回False
    if ifprint:
        print('start download:', url)
    if overwrite: # 如果选择覆盖下载，判断原有文件是否存在，若存在则先删除该文件再尝试下载
        ifexist = os.path.exists(save_name)
        if ifexist:
            os.remove(save_name)
    try:
        if ifbar:
            download_path = wget.download(url, out=save_name)
        else:
            download_path = wget.download(url, out=save_name, bar=None)
        if ifprint:
            print(f'Downloaded: {save_name}') # 下载成功，返回下载文件名称
        return download_path
    except:
        if ifprint:
            print(f'Could not download: {save_name}') # 下载不成功，返回False
        return False

# 测试
#url='https://imslp.eu/files/imglnks/euimg/9/9f/IMSLP708927-PMLP25179-AnAmericanInParis.mp3'
#save_name='./Temp/test.mp3'
#download_url(url, save_name, overwrite=False, ifprint=True, ifbar=True)
```

```python
def csvc(file, col_number, skip_first_row=True): 
    # 输入：csv文件路径，读取第几列，是否跳过首行
    # 返回这一列的数据
    with open(file, encoding='utf-8') as f:
        csvfile = csv.reader(f)
        countr=0
        data=[]
        for row in csvfile: 
            countr+=1 #每读一行+1
            if skip_first_row == True:
                if countr==1: #跳过首行
                    continue
            data.append(row[col_number])    
    return data

# 测试
#file = './Temp/MW2014_8.34_230_expcurve.csv'
#col_number = 2
#csvc(file, col_number)
```

```python
def read_fits_col(file, col_name, hdu_list_number=1): 
    # 输入：文件路径，列数，hdu编号
    # 函数读取对应列的数据，并返回
    hdu = fits.open(file)
    col_data = hdu[hdu_list_number].data.field(col_name)
    hdu.close()
    return col_data
    
# 测试
#file = './LIB/VSX20230729_ra120-240.fits'
#col_name = 'min'
#read_fits_col(file, col_name, hdu_list_number=1)
```

```python
def read_fits_cols(file, col_names, hdu_list_number=1): 
    # 输入：文件路径，列数，hdu编号
    # 函数读取对应列的数据，并返回   
    hdu = fits.open(file)
    col_data = []
    for col_name in col_names:
        col_data.append(hdu[hdu_list_number].data.field(col_name))
    hdu.close()
    return col_data

# 测试
#file = './LIB/VSX20230729_ra120-240.fits'
#col_names = ['_RAJ2000','_DEJ2000']
#read_fits_cols(file, col_names, hdu_list_number=1)
```

```python
def create_dir(directory):
    # 输入：目录路径
    # 如果目录不存在，就新建目录
    if not os.path.exists(directory):
        os.mkdir(directory)
    return directory

# 测试
#directory = 'new_folder'
#create_dir(directory)
```

```python
def get_filelist(path):
    # 输入：目录路径
    # 功能：读取path目录中的所有文件，返回按顺序排列的目录下的文件列表
    return sorted(os.listdir(path)) 

# 测试
#path = 'LIB'
#get_filelist(path)
```

```python
def save_data_totxt(data, save_txt_name='./saved_data.txt'):
    # 输入：data字符串数据，保存文件名称
    # 写入并覆盖txt文件
    with open(save_txt_name,'w',encoding='utf-8') as f:
        f.write(data)
    return save_txt_name

# 测试
#data = 'Hello, world!\nBye, world.'
#file = './Temp/test_save.txt'
#save_data_totxt(data, file)
```

```python
def coord2name(coord):
    # 输入：坐标对象
    # 功能：输出以小时为单位的名称字符串
    string = coord.to_string('hmsdms', sep='', precision=2, alwayssign=True)
    last2 = str(round(float(string[-5:]),1))
    return 'J' + string[:-5][1:].replace(' ','') + last2.rjust(4,'0')

# 测试
#coord = SkyCoord(ra=103.1549367983, dec=+24.60618197635, unit=(u.deg,u.deg), frame='icrs')
#coord = SkyCoord('103.0363 6.96817', unit=(u.deg,u.deg), frame='icrs')
#coord2name(coord)
```
