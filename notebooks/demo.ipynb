# 🔭 Astro-AI: 天文智能分析系统演示

本笔记本演示如何使用Astro-AI系统进行天体智能分析。

## 案例: AM Her (Polar型激变变星)

```python
# 导入必要的库
import sys
sys.path.insert(0, '../src')

from astro_agent import AstroAIAgent
from rag_system import AstronomyRAG
from astro_tools import AstroTools

print("✓ 库导入成功")
```

## 1. 初始化智能体

```python
# 初始化AI智能体
agent = AstroAIAgent()
```

## 2. RAG知识库查询

检索Polar型激变变星的相关知识。

```python
# 创建RAG知识库
rag = AstronomyRAG()

# 查询Polar相关知识
knowledge = rag.search("polar cataclysmic variable")
print(knowledge[:1000])
```

## 3. 天文工具使用

查询AM Her的观测数据。

```python
# 创建工具集
tools = AstroTools()

# AM Her坐标
ra = 274.0554  # 度
dec = 49.8679  # 度

print(f"分析目标: AM Her")
print(f"坐标: RA={ra}°, DEC={dec}°")
print("="*60)
```

```python
# 查询消光
ext = tools.query_extinction(ra, dec)
print("\n【消光】")
print(ext)
```

```python
# 查询测光目录
phot = tools.query_photometry(ra, dec)
print("\n【测光】")
print(phot)
```

## 4. 完整天体分析

使用智能体进行完整的分析流程。

```python
# 分析AM Her
result = agent.analyze_target(
    ra=ra,
    dec=dec,
    name="AM_Her_Polar"
)
```

```python
# 显示分析结果
print("\n" + "="*60)
print("📊 AI分析结果")
print("="*60)
print(result['analysis'])
```

## 5. 结果可视化

```python
import matplotlib.pyplot as plt
import json

# 读取保存的结果
with open('../output/AM_Her_Polar_analysis.json', 'r') as f:
    data = json.load(f)

# 显示数据摘要
print(f"目标: {data['name']}")
print(f"时间: {data['timestamp']}")
print(f"消光 A_V: {data['extinction'].get('A_V', 'N/A')}")
print(f"ZTF数据点: {data['ztf'].get('n_points', 'N/A')}")
```
